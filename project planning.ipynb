(4) Methods and Plan

We propose using linear regression as the model to explore the relationship between these variables

Why is this method appropriate?

Linear regression is appropriate because it allows us to model continuous outcomes as a linear function of predictor variables. Since the duration(one of the uncreated variable, or played hours) is a continuous numerical variable, linear regression will help us identify relationships  how different factors influence session length.

So we can predict what kind of person features have the potential to have more duration in game playing, we can recruit that group of people.

Cross-validation ensures that the model generalizes well to unseen data, reduce the risk of overfitting, which is crucial when working with smaller datasets or when we do not have a lot of data to validate the model's performance effectively.

assumption

Linearity: The relationship between the predictors and the session duration is assumed to be linear.

Potential Limitations 

Overfitting and Non-linear relationships
